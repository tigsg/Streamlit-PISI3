```python
#importando bibliotecas
import pandas as pd
```

```python
#Carregando os dados
dataset = pd.read_parquet("TMDB_movie_dataset_v11.parquet")[:500] #por enquanto só os 500 primeiros registros
```

```python
#removendo os filmes com classificação para adultos
dataset = dataset[dataset['adult'] == False]
```

```python
#criando uma cópia focada nos gêneros para fazer o one-hot encoding
dataset_generos = dataset[['id', 'title', 'genres']].copy()
```

```python
#Encontrando os gêneros de filmes presentes no dataset
generos = dataset_generos['genres'].str.split(', ').explode().unique()
generos = generos[:len(generos) - 1]
```

```python
#criando o one-hot encoding
for genero in generos:
    dataset_generos[genero] = dataset_generos['genres'].apply(lambda x: 1 if genero in x else 0)
```
